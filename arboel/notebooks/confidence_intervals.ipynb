```python
with open("data/medm_results.json", "r") as handle:
    res = handle.read()
```

```python
res = json.loads(res)
```

```python
res
```

```python
len(res['failure'])
```

```python
len(res['success'])
```

```python
success = {obj['mention_id']: 1 for obj in res['success']}
failure = {obj['mention_id']: 0 for obj in res['failure']}
```

```python
merged = {**success, **failure}
```

```python
from scipy.stats import bootstrap
import numpy as np
```

```python
# BCa Bootstrap confidence interval

rng = np.random.default_rng(17)
data = (np.array(list(merged.values())), )
res = bootstrap(data, np.mean, confidence_level=0.9, random_state=rng)
```
