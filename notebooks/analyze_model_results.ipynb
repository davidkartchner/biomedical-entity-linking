```python
PLANT_CATEGORIES = [
    'T096',  # Group
    'T002',  # Plant
]

DISEASE_CATEGORIES = [
    'T020',  # Acquired abnormality
    'T190',  # Anatomical abnormaility
    'T049',  # Cell or Molecular dysfunction
    'T019',  # Congential abnormality
    'T047',  # Disease or Syndrome
    # 'T037',  # Injury or Poising
    'T048',  # Mental or behavioural dysfunction
    'T191',  # Neoplastic process
    'T046',  # Pathologic function
    'T184',  # Sign or Symptom
]

PROTEIN_CATEGORIES = [
    'T192',  # Receptor
    'T126',  # Enzyme
    'T116',  # Amino acid, peptide, protein
]

PROCESS_CATEGORIES = [
    'T038',  # Biological process
    'T043',  # Cell function
    'T044',  # Molecular function
    'T039',  # Physiological function
]

custom_categories = {'plant_categories':PLANT_CATEGORIES, 'disease_categories':DISEASE_CATEGORIES, 'protein_categories':PROTEIN_CATEGORIES, 'process_categories':PROCESS_CATEGORIES}
```

```python
def load_types(types_path='/efs/davidkartchner/2021AB_semantic_network/SRDEF'):
    type_df = pd.read_csv(types_path, usecols=[1,2], names=['tui','type_name'], sep='|')
    type2names = {x[0]:x[1] for x in type_df.values}
    return type2names
```

```python
type2name = load_types()
```

```python
def _hit_at_k(gold_db_id, candidates, k):
    '''
    Calculate if one of the top k ranked entities was the correct one
    '''
    if gold_db_id in [x['db_id'] for x in candidates[:k]]:
        return True
    else:
        return False

def evaluate_linking(preds, type2name, ks=[2**i for i in range(7)], custom_types=None):
    '''
    Evaluate entity linking for "normal" entity linking models (i.e. models that make predictions for every span in the dataset)
    '''
    ks = np.array(ks)
    pred_summary = {semtype: {'pred': np.zeros_like(ks), 'total':np.zeros_like(ks)} for semtype in type2name.keys()}
    if custom_types is not None:
        for key in custom_types.keys():
            pred_summary[key.upper()] = {'pred': np.zeros_like(ks), 'total':np.zeros_like(ks)}
    #     pred_summary['overall'] = {'pred': np.zeros_like(ks), 'total':np.zeros_like(ks)}
    # errors = defaultdict(list)
    # successes = defaultdict(list)
    

    omitted = 0
    for pred_dict in preds:
        sem_type = pred_dict['type']
        gold_db_id = pred_dict['normalized'][0]['db_id']
        candidates = pred_dict['candidates']
        
        orig_semtype = deepcopy(sem_type)
        if custom_types is not None:
            cui = gold_db_id.split(':')[-1]
            if cui not in cui2types:
                omitted += 1
                continue
            
            sem_type_list = cui2types[cui]
            for cat, type_vals in custom_categories.items():
                if any([x in type_vals for x in sem_type_list]):
                    sem_type = cat.upper()
                    break
                    
            if sem_type == orig_semtype:
                continue
                    
        
        pred_summary[sem_type]['pred'] += np.array([_hit_at_k(gold_db_id, candidates, k) for k in ks])
        pred_summary[sem_type]['total'] += np.ones_like(ks)
           
    filtered_summary = {k:v for k, v in pred_summary.items() if v['total'][0] > 0}
    
    print(f"Omitted {omitted} entity mentions because no longer in UMLS")

    return filtered_summary

def format_pred_summary(pred_summary, type2name, ks=[2**i for i in range(7)]):
    '''
    Format summary of predictions
    '''
    overall_key = '**Overall**'
    # Accomodate custom type groupings
    type2self = {key:key for key in pred_summary.keys()}
    expanded_type2name = {**type2self, **type2name}
    
    data = {expanded_type2name[k]: v['pred']/v['total'] for k, v in pred_summary.items()}
    pred_counts =  np.array([x['pred'] for x in pred_summary.values()]).sum(axis=0)
    total_counts =  np.array([x['total'] for x in pred_summary.values()]).sum(axis=0) 
    
    category_totals = {expanded_type2name[key]: val['total'][0] for key, val in pred_summary.items()}
    category_totals[overall_key] = np.sum([x for x in category_totals.values()])
    print(category_totals)
    data[overall_key] = pred_counts/total_counts
    output_df = pd.DataFrame.from_dict(data, orient='index', columns=[f'recall@{k}' for k in ks])
    output_df.index.name = 'Semantic Type'
    output_df['Total'] = output_df.index.map(category_totals)
    return output_df.reset_index()
```

```python
ks = [1, 2, 4, 8, 16, 32]
output = evaluate_linking(preds, type2name, ks=ks, custom_types=custom_categories)
output
```

```python
sapbert_custom_preds = format_pred_summary(output, type2name, ks=ks)
sapbert_custom_preds['Model'] = 'SapBERT'
sapbert_custom_preds
# sapbert_preds.to_csv('sapbert_summary.tsv', sep='\t', index=False)
```
